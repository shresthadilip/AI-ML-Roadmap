
## 1. Introduction
Matplotlib is a low-level Python library for creating static, animated, and interactive visualizations.  
Other visualization libraries (like **Seaborn**, **Pandas**, **Plotly**) build on top of Matplotlib.


```python

import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```


## 2. The Matplotlib Architecture: Figure, Axes, Artist

| Component | Description |
|------------|-------------|
| **Figure** | The entire window or page the plot is drawn on |
| **Axes** | The area containing data (multiple per Figure) |
| **Axis** | X or Y coordinate system line |
| **Artist** | Anything visible on a plot (lines, text, patches, etc.) |


```python

fig, ax = plt.subplots()
ax.plot([1, 2, 3, 4], [10, 20, 25, 30])
ax.set_xlabel("X-Axis")
ax.set_ylabel("Y-Axis")
ax.set_title("Basic Line Plot")
plt.show()
```


## 3. Figures and Subplots
You can create multiple plots within a single figure using `plt.subplots()`.


```python

fig, axes = plt.subplots(1, 2, figsize=(10, 4))

axes[0].plot([1, 2, 3], [1, 4, 9])
axes[0].set_title("Left Plot")

axes[1].plot([1, 2, 3], [1, 2, 3])
axes[1].set_title("Right Plot")

plt.tight_layout()
plt.show()
```


## 4. Common Plot Types


```python

# Line Plot
plt.plot([0, 1, 2, 3], [10, 20, 25, 30], color='blue', linestyle='--', marker='o')
plt.title("Line Plot")
plt.show()

# Scatter Plot
plt.scatter([1, 2, 3, 4], [10, 20, 25, 30], color='red')
plt.title("Scatter Plot")
plt.show()

# Bar Plot
plt.bar(["A", "B", "C"], [10, 15, 7], color='green')
plt.title("Bar Plot")
plt.show()

# Histogram
data = np.random.randn(1000)
plt.hist(data, bins=30, color='purple', edgecolor='black')
plt.title("Histogram")
plt.show()

# Pie Chart
labels = ['A', 'B', 'C']
sizes = [40, 35, 25]
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title("Pie Chart")
plt.show()
```


## 5. Customizing Plots


```python

# Titles and Labels
plt.plot([1, 2, 3], [4, 9, 16])
plt.title("Customized Plot")
plt.xlabel("X Axis Label")
plt.ylabel("Y Axis Label")
plt.show()

# Axes Limits
plt.plot([1, 2, 3], [4, 9, 16])
plt.xlim(0, 5)
plt.ylim(0, 20)
plt.show()

# Legends
plt.plot([1, 2, 3], [1, 4, 9], label='Squared')
plt.plot([1, 2, 3], [1, 8, 27], label='Cubed')
plt.legend()
plt.show()

# Gridlines
plt.plot([1, 2, 3], [1, 4, 9])
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```


## 6. Styling and Saving


```python

# Line Styles and Colors
plt.plot([1, 2, 3], [1, 4, 9], 'r--', label='Dashed Red Line')
plt.legend()
plt.show()

# Save to File
plt.plot([1, 2, 3], [1, 4, 9])
plt.title("Saving Example")
plt.savefig("my_plot.png", dpi=300, bbox_inches='tight')
plt.show()
```


## 7. Object-Oriented Interface (Best Practice)


```python

fig, ax = plt.subplots(figsize=(6,4))
ax.plot([1, 2, 3], [1, 4, 9], label='Squared')
ax.set_title("Object-Oriented Plot")
ax.set_xlabel("X-Axis")
ax.set_ylabel("Y-Axis")
ax.legend()
plt.show()
```
